```python
from keras.preprocessing.image import img_to_array,load_img,ImageDataGenerator
```

```python
datagen=ImageDataGenerator(shear_range=0.5,
                           zoom_range=0.4
                           ,rotation_range=7,
                           width_shift_range=4,
                           height_shift_range=4)
```

```python
img=load_img(r"E:\STIP TASKS\Traffic signals\test_data_traffic\speed_lim_test\speedlimit11.jpg")
```

```python
import numpy as np
i_to_a=img_to_array(img)
reshape_i_to_a=i_to_a.reshape((1,)+i_to_a.shape)
```

```python
i=0
for batch in datagen.flow(reshape_i_to_a,
                          batch_size=1,
                          save_to_dir=r"E:\STIP TASKS\Traffic signals\test_data_traffic\speed_lim_test",
                          save_prefix="speed_lim121",
                          save_format="jpg"):
    i+=1
    if i>20:
        break
```
